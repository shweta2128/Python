```python
import pandas as pd
```

```python
#Q1. How do you load a CSV file into a Pandas DataFrame?
df2 = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df2.head()
```

```python
#Q2. How do you check the data type of a column in a Pandas DataFrame?
df2.dtypes
```

```python
#Q3. How do you select rows from a Pandas DataFrame based on a condition?
#Below will select rows with Pclass = 1
df2[df2['Pclass']==1]
```

```python
#Q4. How do you rename columns in a Pandas DataFrame?
#using rename function , inplace = True will change it in dataframe, in next call dataframe column name will be ‘PID’
df2.rename(columns={"PassengerId":"PID"},inplace=True)
df2.head()
```

```python
#Q5. How do you drop columns in a Pandas DataFrame?
#using drop function , inplace = True will change it in dataframe, in next call dataframe Cabin and Embarked columns will not shown
df2.drop(columns=['Cabin','Embarked'],inplace=True)
df2.head()
```

```python
#Q6. How do you find the unique values in a column of a Pandas DataFrame?
#Below will give unique values in ‘city’ column
df2.Pclass.unique()
```

```python
#Q7. How do you find the number of missing values in each column of a Pandas DataFrame?
df2.isna().sum()
```

```python
#Q8. How do you fill missing values in a Pandas DataFrame with a specific value?
#Below will replace missing value of Age column withmean value of that column
df2.fillna({"Age":df2['Age'].mean()},inplace=True)
df2['Age']
```

```python
#Q9. How do you concatenate two Pandas DataFrames?
#Using concat function, below df3 is concatenation of df1 and df2 dataframes.
df = pd.DataFrame({"A":['a'],"B":['b']})
df1 = pd.DataFrame({"A":['c'],"B":['d']})

df3 = pd.concat([df,df1])
df3
```

```python
#Q10. How do you merge two Pandas DataFrames on a specific column?
#Using merge function, below df3 is result of merge of df1 and df2 dataframes on common column.
df4 = pd.DataFrame({"A":['a'],"B":['b']})

df5 = pd.DataFrame({"B":['b'],"C":['d']})
df6 = pd.merge(df4,df5,on="B")
df6
```

```python
#Q11. How do you group data in a Pandas DataFrame by a specific column and apply an aggregation function?
#Below will give passenger count of each Class in Titanic data
df2.groupby("Pclass")['Pclass'].count()
```

```python
#Q12. How do you pivot a Pandas DataFrame?
df7 = pd.DataFrame({"A":[1,2,3],"B":['A','B','C'],"C":[5,6,4]})

df7.pivot(index='A', columns='B', values='C')
```

```python
#Q13. How do you change the data type of a column in a Pandas DataFrame?
#Using astype function
df2 = df2.astype({'Age':float})
df2.head()
```

```python
#Q14. How do you sort a Pandas DataFrame by a specific column?
#Below will sort titanic data by Age column
df2.sort_values(by='Age')
```

```python
#Q15. How do you create a copy of a Pandas DataFrame?
df1 = df2.copy()
df1.head()
```

```python
#Q16. How do you filter rows of a Pandas DataFrame by multiple conditions?
#Below will give Male passengers with age grater than 40
df2[(df2['Age'] > 40)&(df2['Sex']=='male')]
```

```python
#Q17. How do you calculate the mean of a column in a Pandas DataFrame?
#Below will give mean of Fare in titanic dataframe
df2['Fare'].mean()
```

```python
#Q18. How do you calculate the standard deviation of a column in a Pandas DataFrame?
df2['Fare'].std()
```

```python
#Q19. How do you calculate the correlation between two columns in a Pandas DataFrame?
#Below will calculate correlation between “Pclass” and “Fare” columns of titanic dataset.
df2['Pclass'].corr(df2['Fare'])
```

```python
#Q20. How do you select specific columns in a DataFrame using their labels?
df2['Pclass']
```

```python
#Q21. How do you select specific rows in a DataFrame using their indexes?
#Below will give second row data
df2.iloc[2]
```

```python
#Q22. How do you sort a DataFrame by a specific column?
df2.sort_values(by='Fare')
```

```python
#Q23. How do you create a new column in a DataFrame based on the values of another column?
#Below will add “New_Column” with values addition of SibSp and Parch columns.
df2['New_Column'] = df2['SibSp']+df2['Parch']
df2.head()
```

```python
#Q24. How do you remove duplicates from a DataFrame?

df2.drop_duplicates(subset="Ticket",keep=False, inplace=True)
df2
```

```python
#Q25. What is the difference between .loc and .iloc in Pandas?

The .loc function is label based data selecting method which means that we have to pass the name of the row or column which we want to select.
loc() can accept the boolean data unlike iloc()
The iloc() function is an indexed-based selecting method which means that we have to pass an integer index in the method to select a specific row/column.
iloc() does not accept the boolean data unlike loc().
```
